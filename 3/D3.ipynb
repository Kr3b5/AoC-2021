```python
# P1
f = open("input.txt", "r")

list = f.readlines()
out = []
n1 = 0 
n0 = 0

for x in range(12):
    for y in list:
        if(int(y[x]) == 1):
            n1 = n1 + 1
        else:
            n0 = n0 + 1
    if ( n1 > n0 ):
        out.append(1)
    else:
        out.append(0)
        
    n0 = n1 = 0

gamma = ''.join(str(e) for e in out)

epsilon = ''
for i in gamma:
    if i == '0':
        epsilon += '1'
    else:
        epsilon += '0'

print(int(gamma, 2) * int(epsilon, 2))
```
